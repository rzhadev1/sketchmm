```python
from engine import Engine
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import seaborn as sns
from scipy.stats import maxwell
from util import maxwellboltzmann
import time
sns.set(color_codes=True)
```

```python
e = Engine()
e.load_settings("./settings.yaml")
time_steps = 2000
#e.load_positions("./data.xyz")
```

```python
initpos = e.pos.copy()
# graph the energy/temperature vs time
velocities = np.zeros(e.N)
vx = np.zeros((time_steps, e.N))
vy = np.zeros((time_steps, e.N))
temp = np.zeros((time_steps, 2))
energy = np.zeros((time_steps, 2))
pressure = np.zeros((time_steps, 2))
```

```python
plt.figure(1);
plt.scatter(initpos[:, 0], initpos[:, 1])
plt.savefig('startpos')
```

```python
for i in range(time_steps):
    e.step_forward()
    temp[i] = np.asarray((e.time, e.avgT))
    vx[i] = e.vel[:, 0]
    vy[i] = e.vel[:, 1]
    energy[i] = np.asarray((e.time, e.energy))
    print(e.potential_E)
    pressure[i] = np.asarray((e.time, e.avgP))
endpos = e.pos.copy()
scale = np.sqrt(e.avgT)    
velocities = np.sqrt(e.vel[:, 0] * e.vel[:, 0] + e.vel[:, 1] * e.vel[:, 1])
```

```python
a = np.linspace(0, 3, 1000)
plt.figure(2);
sns.lineplot(a, maxwellboltzmann(a, 1, 1, .5), label="Maxwell Boltzmann PDF")
sns.histplot(velocities, bins=20, hist=True, kde=True, label="Engine")
plt.legend()
plt.title(f"Speed Distribution At T = {e.avgT:.3f}")
plt.xlabel("Velocity")
plt.ylabel("Frequency")
plt.savefig('maxwell')
```

```python
plt.figure(3);
plt.scatter(initpos[:, 0], initpos[:, 1], color="brown", label="Initial Pos")
plt.scatter(endpos[:, 0], endpos[:, 1], color="black", label="Final Pos")
plt.title("Particle Positions")
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.savefig('positions')
```

```python
plt.figure(4);
energy[:, 1] /= e.N
plt.plot(energy[:, 0], energy[:, 1])
plt.xlabel("Engine Time")
plt.ylabel("Energy per Atom")
plt.title('Time vs Energy/Atom')
plt.savefig('energy')
```

```python
plt.figure(5);
plt.plot(temp[4:, 0], temp[4:, 1])
plt.xlabel("Engine Time")
plt.ylabel("Time-Avg Temperature")
plt.title('Time vs Temperature Avg')
plt.savefig('temp')
```

```python
plt.figure(6);
plt.plot(pressure[10:, 0], pressure[10:, 1])
plt.xlabel("Engine Time")
plt.ylabel("Time-Avg Pressure")
plt.title('Time vs Pressure Avg')
plt.savefig('pressure')
```
